### Why is Proof Necessary?

Proofs play a vital role in confirming whether a statement or theorem is true. They provide logical reasoning that demonstrates the validity of a claim, ensuring its reliability under specific circumstances. This reliability enables us to use the statement confidently in various calculations, deductions, or practical applications.

    
### Consequences of Failing to Provide a Proof

Without a proof, we can't be sure if a statement or theorem is true. This uncertainty makes it hard to know if the claim applies in different situations. So, the statement stays uncertain, making it tough to use in solving problems or understanding things better.

# 1. Prove that the Euler's number $e = \Sigma_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$.
---
Recall that a rational number is one that can be expressed as the quotient of two integers. We want to prove that Euler's number $e = \sum_{k=0}^\infty \frac{1}{k!} \notin \mathbb{Q}$.

**Proof by Contradiction**

We will assume the opposite (that `e` is rational) and show that this leads to a contradiction.

* **Definition of Rational Numbers:**

A rational number is any number that can be expressed as a quotient (`a / b`), where `a` and `b` are integers, and `b ≠ 0`.

* **Assume e is Rational:**

For the sake of contradiction, let's assume `e` can be expressed as a fraction `a / b`, where `a` and `b` are integers.

* **Series Expansion of e:**

We know the series expansion of `e`:

$$ e = \sum_{k=0}^\infty \frac{1}{k!} = 1 + \frac{1}{1!} + \frac{1}{2!} + \frac{1}{3!} + \dotsb $$

* **Multiplying by b!**

Let's multiply both sides of the equation by `b!` to eliminate the factorials:

$$ b!e = b! \left(1 + \frac{1}{1!} + \frac{1}{2!} + \frac{1}{3!} + \dotsb \right) = b! (1 + b + \frac{b(b-1)}{2!} + \frac{b(b-1)(b-2)}{3!} + \dotsb) $$

* **Contradiction:**

Observe that every term on the right side of the equation is now a multiple of `b!` because it includes `b!` or a product of factors containing `b`. Since the product of integers is an integer, the entire right side becomes an integer. This contradicts our initial assumption that `e = a / b` (where `b` is an integer). 

Therefore, our assumption that `e` is rational must be false. 

**Conclusion:**

By contradiction, we have proven that Euler's number `e` is irrational (i.e., it cannot be expressed as a fraction of two integers).

Observe that since each term on the right side of the equation is a multiple of $b!$, the entire expression is an integer. Consequently, if $e$ were rational, it would contradict our initial assumption. Thus, $e \notin \mathbb{Q}$.



## 2. Prove Minkowski's Inequality for sums

**Theorem:** For all positive real numbers $p > 1$ and sequences $(a_k, b_k) > 0$, the following inequality holds:

$$ \left(\sum_{k=1}^n |a_k + b_k|^p\right)^\frac{1}{p} \leq \left(\sum_{k=1}^n |a_k|^p\right)^\frac{1}{p} + \left(\sum_{k=1}^n |b_k|^p\right)^\frac{1}{p} $$

**Proof:**

The proof relies on Hölder's inequality. Let $q$ be the conjugate exponent of $p$ (i.e., $\frac{1}{p} + \frac{1}{q} = 1$).

We apply Hölder's inequality with exponents $p$ and $q$ to two pairs of sequences:

1. $(|a_k|^p, 1)$:

$$ \sum_{k=1}^n |a_k|^p \cdot 1 \leq \left(\sum_{k=1}^n |a_k|^p\right)^\frac{1}{p} \left(\sum_{k=1}^n 1^q\right)^\frac{1}{q} = \left(\sum_{k=1}^n |a_k|^p\right)^\frac{1}{p} \cdot n^\frac{1}{q} $$

2. $(|b_k|^p, 1)$:

$$ \sum_{k=1}^n |b_k|^p \cdot 1 \leq \left(\sum_{k=1}^n |b_k|^p\right)^\frac{1}{p} \left(\sum_{k=1}^n 1^q\right)^\frac{1}{q} = \left(\sum_{k=1}^n |b_k|^p\right)^\frac{1}{p} \cdot n^\frac{1}{q} $$

Adding these two inequalities:

$$ \sum_{k=1}^n |a_k|^p + \sum_{k=1}^n |b_k|^p \leq \left(\sum_{k=1}^n |a_k|^p\right)^\frac{1}{p} \cdot n^\frac{1}{q} + \left(\sum_{k=1}^n |b_k|^p\right)^\frac{1}{p} \cdot n^\frac{1}{q} $$

Since all terms are positive, we can rearrange:

$$ \sum_{k=1}^n (|a_k|^p + |b_k|^p) \leq \left(\sum_{k=1}^n |a_k|^p\right)^\frac{1}{p} \cdot n^\frac{1}{q} + \left(\sum_{k=1}^n |b_k|^p\right)^\frac{1}{p} \cdot n^\frac{1}{q} $$

Taking the $p^{th}$ root of both sides:

$$ \left(\sum_{k=1}^n (|a_k|^p + |b_k|^p)\right)^\frac{1}{p} \leq \left( \left(\sum_{k=1}^n |a_k|^p\right)^\frac{1}{p} \cdot n^\frac{1}{q} + \left(\sum_{k=1}^n |b_k|

^p\right)^\frac{1}{p} \cdot n^\frac{1}{q} \right)^\frac{1}{p} $$

By the AM-GM inequality (which applies since all terms are positive), the right-hand side simplifies to:

$$ \left(\sum_{k=1}^n (|a_k|^p + |b_k|^p)\right)^\frac{1}{p} \leq \left(\sum_{k=1}^n |a_k|^p\right)^\frac{1}{p} + \left(\sum_{k=1}^n |b_k|^p\right)^\frac{1}{p} $$

This completes the proof of Minkowski's Inequality for sums.


# 3. Prove the triangle inequality $|x+y| \leq |x| + |y|, \forall (x,y) \in \mathbb{R}$
---

$\textit{Proof:}$

Consider three cases:

1. If $x \geq 0$ and $y \geq 0$, then $|x| = x$ and $|y| = y$, so $|x+y| = x+y$. Then $|x+y| = x + y \leq |x| + |y|$.
2. If $x \geq 0$ and $y < 0$, then $|x| = x$ and $|y| = -y$, so $|x+y| = x-y$. Then $|x+y| = x - y \leq |x| + |y|$.
3. If $x < 0$ and $y < 0$, then $|x| = -x$ and $|y| = -y$, so $|x+y| = -(x+y) = -x - y$. Then $|x+y| = -(x + y) \leq |x| + |y|$.

# 4. Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:
---

Prove Sedrakayan's Lemma $\forall u_i, v_i \in \mathbb{R}^+$:

$\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{(u_i)^2}{v_i}$

$\textit{Proof:}$

Let $\mathbf{u} = (u_1, u_2, ..., u_n)$ and $\mathbf{v} = (v_1, v_2, ..., v_n)$ be vectors in $\mathbb{R}^+$.

By Cauchy-Schwarz Inequality, we have:

$\left(\sum_{i=1}^n u_i v_i\right)^2 \leq \left(\sum_{i=1}^n u_i^2\right) \left(\sum_{i=1}^n v_i^2\right)$

Since $v_i > 0$ for all $i$, we can divide both sides by $\left(\sum_{i=1}^n v_i^2\right): \frac{\left(\sum_{i=1}^n u_i v_i\right)^2}{\left(\sum_{i=1}^n v_i^2\right)} \leq \sum_{i=1}^n u_i^2$

Now, notice that $\left(\sum_{i=1}^n u_i v_i\right) = \mathbf{u} \cdot \mathbf{v}$ and $\sum_{i=1}^n u_i^2 = \|\mathbf{u}\|^2$, where $\|\cdot\|$ denotes the Euclidean norm.

Thus, we have: $\frac{(\mathbf{u} \cdot \mathbf{v})^2}{\|\mathbf{v}\|^2} \leq \|\mathbf{u}\|^2$

Dividing both sides by $\|\mathbf{v}\|^2$, we get: $\frac{(\mathbf{u} \cdot \mathbf{v})^2}{\|\mathbf{v}\|^2 \|\mathbf{v}\|^2} \leq \frac{\|\mathbf{u}\|^2}{\|\mathbf{v}\|^2}$

Simplifying, we obtain: $\frac{(\mathbf{u} \cdot \mathbf{v})^2}{\|\mathbf{v}\|^4} \leq \frac{\|\mathbf{u}\|^2}{\|\mathbf{v}\|^2}$

Now, substituting $\|\mathbf{v}\|^2 = \sum_{i=1}^n v_i$ and rearranging, we get: $\frac{(\sum_{i=1}^n u_i)^2}{\sum_{i=1}^n v_i} \leq \sum_{i=1}^n \frac{u_i^2}{v_i}$

This proves Sedrakayan's Lemma.
